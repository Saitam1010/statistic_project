GDELT Event Files 2019 herunterladen und extrahieren

```python
#Setup & Imports

import os
import requests
from datetime import date, timedelta
from pathlib import Path
```

```python
#Zielordner definieren & automatisch erstellen

BASE_PATH = Path("..")

ZIP_DIR = BASE_PATH / "Rohdaten_zip"
ZIP_DIR.mkdir(parents=True, exist_ok=True)

print(f"ZIP-Zielordner: {ZIP_DIR.resolve()}")
```

```python
#GDELT-URL & Zeitraum definieren 01.01.2019 – 31.12.2019

GDELT_BASE_URL = "http://data.gdeltproject.org/events"

start_date = date(2019, 1, 1)
end_date   = date(2019, 12, 31)
```

```python
#Download-Funktion

def download_file(url, target_path):
    if target_path.exists():
        print(f"✔ Bereits vorhanden: {target_path.name}")
        return

    print(f"↓ Lade herunter: {target_path.name}")
    response = requests.get(url, stream=True)

    if response.status_code != 200:
        print(f"⚠ Fehler ({response.status_code}) bei {url}")
        return

    with open(target_path, "wb") as f:
        for chunk in response.iter_content(chunk_size=8192):
            f.write(chunk)
```

```python
#Download-Schleife für ganzes Jahr 2019

current_date = start_date

while current_date <= end_date:
    date_str = current_date.strftime("%Y%m%d")
    filename = f"{date_str}.export.CSV.zip"
    url = f"{GDELT_BASE_URL}/{filename}"

    target_file = ZIP_DIR / filename
    download_file(url, target_file)

    current_date += timedelta(days=1)
```

```python
#Kontrolle auf Vollständigkeit-> sollten 365 Dateien sein

zip_files = list(ZIP_DIR.glob("*.zip"))
print(f"Anzahl ZIP-Dateien: {len(zip_files)}")
```

```python
# Setup & Imports

import zipfile
from pathlib import Path
```

```python
# Definition des Basis-Pfads (Projekt-Root, eine Ebene über diesem Notebook)
BASE_PATH = Path("..")

# Definition der Ordner für ZIP-Dateien und extrahierte Rohdaten
ZIP_DIR = BASE_PATH / "Rohdaten_zip"
EXTRACT_DIR = BASE_PATH / "Rohdaten_extrahiert"
```

```python
# Erstellen des Zielordners für die extrahierten CSV-Dateien
EXTRACT_DIR.mkdir(parents=True, exist_ok=True)

print(f"ZIP-Ordner: {ZIP_DIR.resolve()}")
print(f"Extrahierter Datenordner: {EXTRACT_DIR.resolve()}")
```

```python
# Auflisten aller ZIP-Dateien im Rohdaten_zip-Ordner
zip_files = sorted(ZIP_DIR.glob("*.zip"))

print(f"{len(zip_files)} ZIP-Dateien gefunden")
```

```python
# Iteratives Entpacken aller ZIP-Dateien in den Zielordner
for zip_path in zip_files:
    try:
        with zipfile.ZipFile(zip_path, "r") as zf:
            zf.extractall(EXTRACT_DIR)
        print(f"✔ extrahiert: {zip_path.name}")
    except Exception as e:
        print(f"✖ Fehler bei {zip_path.name}: {e}")
```

```python
# Kontrolle auf Vollständigkeit der extrahierten CSV-Dateien

csv_files = list(EXTRACT_DIR.glob("*.CSV"))
print(f"Extrahierte CSV-Dateien: {len(csv_files)}")
```
